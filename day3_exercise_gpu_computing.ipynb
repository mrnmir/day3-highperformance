```python
import numpy as np
import cupy as cp
```

```python
n = 2048
```

```python
%%timeit
np_array = np.random.rand(n, n).astype(np.float32)
np.fft.fft2(np_array)
```

```python
%%timeit
cp_array = cp.random.rand(n, n)
cp.fft.fft2(cp_array)
```
